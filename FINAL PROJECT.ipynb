```python
class FoodItem:
    def __init__(self, food_id, name, quantity, price, discount, stock):
        self.food_id = food_id
        self.name = name
        self.quantity = quantity
        self.price = price
        self.discount = discount
        self.stock = stock

class User:
    def __init__(self, full_name, phone_number, email, address, password):
        self.full_name = full_name
        self.phone_number = phone_number
        self.email = email
        self.address = address
        self.password = password
        self.orders = []

    def place_order(self, food_items):
        order = Order(self, food_items)
        self.orders.append(order)
        print("Order placed successfully!")

class Order:
    def __init__(self, user, food_items):
        self.user = user
        self.food_items = food_items

    def calculate_total(self):
        total = sum(item.price * item.quantity for item in self.food_items)
        return total

    def display_order(self):
        print("Order Details:")
        for item in self.food_items:
            print(f"{item.name} ({item.quantity}) - ${item.price * item.quantity}")
        print(f"Total: ${self.calculate_total()}")

def main():
    # Sample food items
    food_items = [
        FoodItem(1, "Tandoori Chicken", "4 pieces", 240, 0, 10),
        FoodItem(2, "Vegan Burger", "1 piece", 320, 0, 20),
        FoodItem(3, "Truffle Cake", "500gm", 900, 0, 5),
    ]

    # Sample user
    user = User("Name", "123-456-7890", "hjauabattula21@gmail.com", "123 Main St", "password")

    while True:
        print("\nOptions:")
        print("1. Place New Order")
        print("2. View Order History")
        print("3. Update Profile")
        print("4. Exit")
        choice = input("Enter your choice: ")

        if choice == "1":
            print("Available Food Items:")
            for item in food_items:
                print(f"{item.food_id}. {item.name} ({item.quantity}) - ${item.price}")
            
            selected_items = []
            item_ids = input("Enter item numbers to order (comma-separated): ").split(',')
            for item_id in item_ids:
                item_id = int(item_id)
                if 1 <= item_id <= len(food_items):
                    selected_items.append(food_items[item_id - 1])
                else:
                    print("Invalid item number.")
            user.place_order(selected_items)

        elif choice == "2":
            if user.orders:
                print("Order History:")
                for order in user.orders:
                    order.display_order()
            else:
                print("No order history.")

        elif choice == "3":
            print("Update Profile:")
            user.full_name = input("Full Name: ")
            user.phone_number = input("Phone Number: ")
            user.email = input("Email: ")
            user.address = input("Address: ")
            user.password = input("Password: ")
            print("Profile updated successfully!")

        elif choice == "4":
            break

if __name__ == "__main__":
    main()
```
